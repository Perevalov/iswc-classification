```python
from google.cloud import translate_v2 as translate
translate_client = translate.Client()
```

```python
TARGET_LANG_LIST = ['de', 'fr', 'es', 'ru']
```

```python
def translate(text, target_lang='ru'):
    if isinstance(text, six.binary_type):
        text = text.decode('utf-8')
    
    result = translate_client.translate(
        text, source_language='en', target_language=target_lang)
    
    return result['translatedText']
```

```python
dbpedia_json = None

with open('../datasets/DBpedia/smarttask_dbpedia_train.json') as json_file:
    dbpedia_json = json.load(json_file)
```

```python
dbpedia_json[0]
```

```python
for i in tqdm.tqdm(range(240, len(dbpedia_json))):
    if dbpedia_json[i]['question']:
        for lang in TARGET_LANG_LIST:
            result = translate(dbpedia_json[i]['question'], lang)
            dbpedia_json[i]['question_{0}'.format(lang)] = result
```
